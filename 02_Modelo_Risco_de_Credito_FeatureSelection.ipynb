* Aplicar one-hot encoding para converter variáveis categóricas em uma representação numérica adequada.

* Separar o target das variáveis para treinar o modelo com RF e obter a importância da variáveis

```python
x = abt_00.drop(axis=1,columns='TARGET')
# Realize a codificação one-hot nas variáveis categóricas
x_encoded = pd.get_dummies(x)

y = abt_00['TARGET']

# Treinar modelo
clf = RandomForestClassifier()
clf.fit(x_encoded, y)
```

```python
# Importância das variáveis

feature_importances = clf.feature_importances_
features = pd.DataFrame({
      'Feature': x_encoded.columns,
      'Importance': feature_importances
})

# Ordenar variáveis por importância
features = features.sort_values(by='Importance', ascending=False)

```
